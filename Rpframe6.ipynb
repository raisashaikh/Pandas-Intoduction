```python
#Splitting or Aggregations or Fiteration or Transformation
import pandas as pd

ipl_data = {'Team': ['MI', 'MI', 'Devils', 'Devils', 'Kings',
   'CSK', 'CSK', 'CSK', 'MI', 'Royals', 'Royals', 'MI'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()# when you want to covert categorical column in numeric data type we can use labelEncoder
```

```python
le.fit_transform(df.Team)#Fit label encoder and return encoded labels.
```

# Transformation calling a function over the selected data

```python
df.groupby('Year')['Points'].transform(lambda x:x+10)
```

```python
def add(x):
    return x+10
df.groupby('Year')['Points'].transform(add)
```

```python
df.Points+100
```

```python
grouped = df.groupby('Team')
score = lambda x: (x + 500)
grouped['Points'].transform(score)
```
